Q3

```python
from scipy.stats import norm,poisson
import numpy as np 
  
data_start = -5
data_end = 5
data_points = 11
data = np.linspace(data_start, data_end, data_points) 
  
mean = np.mean(data) 
std = np.std(data) 
  
probability_pdf = norm.pdf(3, loc=mean, scale=std) 
print(probability_pdf)
```

Q4

A binomial experiment is one that has the following properties: 
1. The experiment consists of n identical trials. 
2. Each trial results in one of the two outcomes, called a success S and failure F.
3. The probability of success on a single trial is equal to p and remains the same from trial to trial.

Eg. Tossing a coin 10 times.
    Shooting a target 10 times.

Binomial distribution describes the distribution of binary data from a finite sample. Thus it gives the probability of getting r events out of n trials. Poisson distribution describes the distribution of binary data from an infinite sample. Thus it gives the probability of getting r events in a population.
