```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure,output_file,show
from bokeh.sampledata.iris import flowers
```

```python
flowers
```

Q1

```python

output_file('test1_html')
p=figure(title='test flower')
p.xaxis.axis_label="petal length"
p.yaxis.axis_label="petal width"
p.circle(flowers['petal_length'], flowers['petal_width'])
show(p)
```

```python

#creating a circle_x glyyph
output_file("circle.html")
p=figure(title='Glyph')
p.circle_x(x=[4, 7, 2, 6], y=[7, 2, 4, 9], size=25,
           fill_color="green", fill_alpha=0.6, line_width=2)
show(p)
```

```python
#Creating a line glyph
x=[3,4,5,6,7]
y=[4,5,6,7,8]
output_file("line1.html")
p=figure(title='line graph')
p.scatter(x,y,size=20,fill_color='red',legend_label='This is my data')
show(p)
```

```python
output_file('test1_html')
p=figure(title='test flower')
p.xaxis.axis_label="petal length"
p.yaxis.axis_label="petal width"
p.line(flowers['petal_length'], flowers['petal_width'],legend_label='This is my data')
show(p)
```
