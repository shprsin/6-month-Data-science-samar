```python
# An eigenvector v of a linear transformation T is a nonzero vector that, when T is applied to it, does not change direction. 
# Applying T to the eigenvector only scales the eigenvector by the scalar value λ, called an eigenvalue. This condition can be 
# written as the equation Av =λv referred to as the eigenvalue equation or eigenequation. In general, λ may be any scalar. For example,
# λ may be negative, in which case the eigenvector reverses direction as part of the scaling, or it may be zero or complex.
```
