```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.DataFrame({'Time_spent':np.random.randint(1,10,5),
                'Final Score':np.random.randint(70,100,5)})
```

```python
df.corr(method='pearson')
```

```python
# There is a Positive relationship between Time Spent and Final Score meaning they increase and decrease with each other. 
```
